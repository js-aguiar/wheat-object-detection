## Training EfficientDet D6 with GWD dataset

This notebook trains an object detection model to find/count wheat heads on aerial images using standard bounding box annotation procedure.

- Set the model and configurations
- Load pre-trained weights
- Train with wheat images and augmentation

The following augmentation techniques are being used:

- Cutout, Random Crop and Flip
- Change brightness, saturation, color and blur
- Cutmix technique
- Mixup technique

```python
import torch
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import source.train as train
from source.dataset import DatasetRetriever
from source.dataset import get_train_transforms, get_valid_transforms
from source.utils import xyxy_to_xywh, plot_image
from source.utils import set_environment_configs
set_environment_configs(seed=10)

class GlobalConfig:
    # Configuration for all the training process
    n_epochs = 2 #60
    num_workers = 4
    batch_size = 1
    lr = 0.00004
    image_height = 512
    image_width = 512
    verbose = True
    
    image_dir = 'data/train'
    boxes_path = 'data/gwd_boxes.csv'
    weights_path = 'data/efficientdet_d6-51cb0132.pth'
    efficientdet_config = 'tf_efficientdet_d6'

    step_scheduler = False
    validation_scheduler = True
    SchedulerClass = torch.optim.lr_scheduler.ReduceLROnPlateau
    scheduler_params = dict(mode='min', factor=0.25, patience=2,
                            verbose=False, threshold=0.0001, threshold_mode='abs',
                            cooldown=0, min_lr=1e-8, eps=1e-08)
```

```python
df = pd.read_csv(GlobalConfig.boxes_path)
df = xyxy_to_xywh(df)
df.head()
```

Split a stratified validation sample and create data retriever objects:

```python
# get unique image ids and their source
images = pd.DataFrame({'image_id': df.image_id.unique()})
source = df[['image_id', 'source']].drop_duplicates(subset=['image_id'])
images = images.merge(source, how='left')

# split train and validation and create dataset
tr, vl = train_test_split(images,test_size=0.1, stratify=images.source)
tr_transform = get_train_transforms(width=GlobalConfig.image_width,
                                 height=GlobalConfig.image_height)
train_dataset = DatasetRetriever(df, tr.image_id.values, GlobalConfig.image_dir,
                                 transforms=tr_transform)
vl_transform = get_valid_transforms(width=GlobalConfig.image_width,
                                 height=GlobalConfig.image_height)
valid_dataset = DatasetRetriever(df, vl.image_id.values, GlobalConfig.image_dir,
                                 test=True, transforms=vl_transform)

print("Train length: {}, val length: {}".format(len(train_dataset), len(valid_dataset)))
```

```python
# a sample image with mixup and cutout augmentation
plot_image(1, train_dataset)
```

```python
# another one with cutmix technique this time
plot_image(1, train_dataset)
```

```python
from effdet import get_efficientdet_config, EfficientDet, DetBenchTrain
from effdet.efficientdet import HeadNet

config = get_efficientdet_config(GlobalConfig.efficientdet_config)
net = EfficientDet(config, pretrained_backbone=False)
net.load_state_dict(torch.load(GlobalConfig.weights_path))
config.num_classes = 1
config.image_size = GlobalConfig.image_width
net.class_net = HeadNet(config, num_outputs=config.num_classes,
                            norm_kwargs=dict(eps=.001, momentum=.01))

net = DetBenchTrain(net, config)
```

```python
train.run_training(train_dataset, valid_dataset, net, GlobalConfig)
```
